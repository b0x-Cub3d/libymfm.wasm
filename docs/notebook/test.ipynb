```rust
:dep plotters = { git = "https://github.com/38/plotters", default_features = false, features = ["evcxr", "line_series", "point_series"] }
extern crate plotters;
use plotters::prelude::*;
use plotters::series::*;

use std::f32::consts::PI;

let figure = evcxr_figure((640, 640), |root| {
    root.fill(&WHITE);
    let mut chart = ChartBuilder::on(&root)
        .caption("ChartBuilder test", ("Arial", 20).into_font())
        .margin(5)
        .x_label_area_size(30)
        .y_label_area_size(30)
        .build_cartesian_2d(-1f32..1f32, -1f32..1f32)?;

    chart.configure_mesh().draw()?;

    chart.draw_series(LineSeries::new(
        (0_f32..2_f32 * PI).step(0.1).values().map(|r| (r.cos(), r.sin())),
        &RED,
    ));

    chart.draw_series(PointSeries::of_element(
        (0_f32..2_f32 * PI).step(0.1).values().map(|r| (r.cos(), r.sin())),
        4,
        ShapeStyle::from(&RED).filled(),
        &|coord, size, style| {
            EmptyElement::at(coord)
                + Circle::new((0, 0), size, style)
                + Text::new(format!("{:?}", coord), (0, 8), ("sans-serif", 8))
        },
    ))?;

    chart.configure_series_labels()
        .background_style(&WHITE.mix(0.8))
        .border_style(&BLACK)
        .draw()?;
    Ok(())
});
figure
```
