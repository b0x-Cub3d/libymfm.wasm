# upsampling

```rust
:dep plotters = { version = "0.3.1", default_features = false, features = ["evcxr", "line_series", "point_series"] }
// :dep libymfm = { path = "/home/hiromasa/devel/amd64/libymfm.wasm" }

extern crate plotters;
// extern crate libymfm;

use plotters::prelude::*;
use plotters::series::*;
use std::f32::consts::PI;

// use libymfm::sound::SoundStream;
// let _stream = SoundStream::new(44100, 96000);

let figure = evcxr_figure((1280, 600), |root| {

    let sampling_rate = 96000_f32;
    let length = sampling_rate / 440_f32;

    root.fill(&WHITE);
    let mut chart = ChartBuilder::on(&root)
        .caption("UpSampling", ("Arial", 20).into_font())
        .margin(5)
        .x_label_area_size(30)
        .y_label_area_size(30)
        .build_cartesian_2d(0f32..length, -1f32..1f32)?;

    chart.configure_mesh().draw()?;

    let rate = 440_f32 * 2_f32;
    let sample_22020: Vec<(f32, f32)> = (0_f32..(22010_f32 / 440_f32)).step(1_f32).values().map(|n| (n, 0.8_f32 * ((2.0_f32 * PI * rate * n) / 22020_f32).sin())).collect();
    let sample_96000: Vec<(f32, f32)> = (0_f32..length).step(1_f32).values().map(|n| (n, 0.8_f32 * ((2.0_f32 * PI * rate * n) / 96000_f32).sin())).collect();

    chart.draw_series(LineSeries::new(
        sample_22020.clone(),
        &RED,
    ))?.label("22020Hz").legend(|(x, y)| PathElement::new(vec![(x, y), (x + 20, y)], &RED));

    chart.draw_series(LineSeries::new(
        sample_96000.clone(),
        &BLUE,
    ))?.label("96000Hz").legend(|(x, y)| PathElement::new(vec![(x, y), (x + 20, y)], &BLUE));

    chart.configure_series_labels()
        .background_style(&WHITE.mix(0.8))
        .border_style(&BLACK)
        .draw()?;
    Ok(())
});
figure
```
